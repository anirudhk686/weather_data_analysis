```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime 
import os
%matplotlib inline

df_1=pd.read_csv(os.getcwd()+'/data/iig_maitri2012-15.csv')
df_1.iloc[:,0]
index=pd.to_datetime(df_1.iloc[:,0])
index
df = pd.DataFrame(data=df_1.values[:,1:],index=index,columns=['t','rh','ws','wd','ap'])
idx = pd.date_range('2012-01-01 00:00:00','2015-12-31 23:00:00',freq='H')
df=df.reindex(idx,fill_value=None)
#filling none values
for col in df:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df = df.interpolate()

x=df['2012-01':'2015-12']

#movingaverage using rolling funct
z=x.iloc[:,0].rolling(720,center=True).mean()

plt.figure(figsize=(30,20))
plt.plot(x.index,x.iloc[:,0]) 
plt.plot(x.index,z,color = 'r',linewidth=3)
plt.ylabel('temperature',fontsize=30)
plt.title('TEMPERATURE WITH ROLLING MEAN - MAITRI',fontdict = {'fontweight': 'bold', 'color': 'red','size':30} )

plt.show()  



```

```python

#movingaverage using rolling funct

z=x.iloc[:,1].rolling(720,center=True).mean()

plt.figure(figsize=(30,20))
plt.plot(x.index,x.iloc[:,1]) 
plt.plot(x.index,z,color = 'r',linewidth=3)

plt.ylabel('relative humidity',fontsize=30)
plt.title('RELATIVE HUMIDITY  - MAITRI',fontdict = {'fontweight': 'bold', 'color': 'red','size':30} )

plt.show()  
```

```python

#movingaverage using rolling funct
z=x.iloc[:,4].rolling(720,center=True).mean()
plt.figure(figsize=(30,20))
plt.plot(x.index,x.iloc[:,4]) 

plt.plot(x.index,z,color = 'r',linewidth=3)
ax = plt.gca()
ax.set_ylim(bottom=940)
plt.ylabel('air pressure',fontsize=30)
plt.title('WINDSPEED - MAITRI',fontdict = {'fontweight': 'bold', 'color': 'red','size':30} )

plt.show()  
```
